En esta clase estaremos aprendiendo a utilizar las graficas en python utilizando la libreira matplotlib(anteriormente necesitas haber instalado la libreria)

```python
import numpy as np
import matplotlib.pyplot as plt
```

Una vez que hayas importado las librerias que usaras, deberas, antes que nada, obtener los datos con los que estaras representando tu grafica

Nota: recuerda que tus arreglos deben tener las mismas dimenciones

```python
x = np.array([0, 10])
y = np.array([0,500])
```

Una vez que tengas los datos que quieres representar, podras modificar las propiedades de la grafica que deseas crear

```python
plt.suptitle("Titulo superior")
plt.title("Titulo")
plt.plot(x, y)
plt.xlabel("Eje x")
plt.ylabel("Eje Y")
plt.show()
```
