### IceCreamData

Dentro de este dataset se encuentran 2 columnas, la primera son los valores de temperatura promedio en un dia y la segunda contiene los ingresos recibidos en ese mismo dia.

```python
# importamos las librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# cargamos el data frame 
df = pd.read_csv("http://virtualfif.uaq.mx/diplomado/data/practica_1/IceCreamData.csv")
```

```python
# analizamos los datos existentes
df.head()
```

```python
# creamos un plot para poder visualizar los datos antes de procesarlos
plt.title("Ice Cream Data")
plt.scatter(df["Temperature"],df["Revenue"])
plt.ylabel("Revenue")
plt.xlabel("Temperature")
plt.grid(1)
plt.show()
```

```python
# hacemos la seleccion de datos y preprocesado
X = df.iloc[:,:-1].values
y = df.iloc[:,1].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=1/5,random_state=0)
```

```python
# entrenamos el modelo
X_train.size
modelo = LinearRegression()
modelo.fit(X_train,y_train)
modelo.fit(X_train,y_train)
y_pred = modelo.predict(X)
```

```python
# mostramos la regresion lineal obtenida
plt.title("Ice Cream Linear Regression")
plt.scatter(df["Temperature"],df["Revenue"])
plt.plot(X,y_pred, color = "red")
plt.ylabel("Revenue")
plt.xlabel("Temperature")
plt.grid(1)
plt.show()
```
