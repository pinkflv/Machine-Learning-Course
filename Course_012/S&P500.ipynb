### S&P500

Ejercicio de pronostico para la S&P500

```python
# importamos las librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# cargamos el data frame 
df = pd.read_csv("http://virtualfif.uaq.mx/diplomado/data/regresion/Multi/S&P500_Stock_Data.csv")
```

```python
# analizamos los datos existentes
df.head()
```

```python
# creamos un plot para poder visualizar los datos antes de procesarlos
plt.suptitle("S&P500")
plt.scatter(df["Interest Rates"],df["S&P 500 Price"])
plt.xlabel("Interest Rates")
plt.ylabel("S&P 500")
plt.grid(1)
plt.show()
```

```python
# creamos un plot para poder visualizar los datos antes de procesarlos
plt.suptitle("S&P500")
plt.scatter(df["Employment"],df["S&P 500 Price"])
plt.xlabel("Employment")
plt.ylabel("S&P 500")
plt.grid(1)
plt.show()
```

```python
# hacemos la seleccion de datos y preprocesado
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=20/100, random_state=666)

modelo = LinearRegression()
```

```python
# Entrenamos el modelo
modelo.fit(X_train,y_train)
```

```python
# Evaluamos el nivel del modelo
modelo.score(X_test,y_test)
```

```python
# Usamos el pronostico para implementarlo
y_pred = modelo.predict(X_train)
```

```python
plt.suptitle("Prediction: Interest")
plt.scatter(X_train[:,0],y_train)
plt.scatter(X_train[:,0],y_pred, color="red")
plt.xlabel("Interest Rates")
plt.ylabel("S&P500")
plt.grid(1)
plt.show()
```

```python
plt.suptitle("Prediction: Employment")
plt.scatter(X_train[:,1],y_train)
plt.scatter(X_train[:,1],y_pred, color="red")
plt.xlabel("Employment")
plt.ylabel("S&P500")
plt.grid(1)
plt.show()
```
