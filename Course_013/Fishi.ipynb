```python
# importamos las librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# cargamos el data frame 
df = pd.read_csv("http://virtualfif.uaq.mx/diplomado/data/regresion/Multi/Fish.csv")
```

```python
# analizamos los datos existentes
df.head()
```

```python
df.info()
```

```python
# creamos un plot para poder visualizar los datos antes de procesarlos
plt.suptitle("Fish Width-Weight")
plt.scatter(df["Width"], df["Weight"])
plt.xlabel("Width")
plt.ylabel("Weight")
plt.grid(1)
plt.show()
```

```python
# creamos un plot para poder visualizar los datos antes de procesarlos
plt.suptitle("Fish Height-Weight")
plt.scatter(df["Height"], df["Weight"])
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid(1)
plt.show()
```

```python
# hacemos la seleccion de datos y preprocesado
X = df.iloc[:, 2:7].values
y = df.iloc[:,1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=3/10, random_state=666)

modelo = LinearRegression()
```

```python
# Entrenamos el modelo
modelo.fit(X_train,y_train)
```

```python
# Evaluamos el nivel del modelo
modelo.score(X_test,y_test)
```

```python
# Usamos el pronostico para implementarlo
y_pred = modelo.predict(X_train)
```

```python
plt.suptitle("Fish Weight")
plt.scatter(X_train[:,4],y_train)
plt.scatter(X_train[:,4],y_pred, color="red")
plt.xlabel("Vol")
plt.ylabel("Weight")
plt.grid(1)
plt.show()
```

```python
plt.suptitle("Fish Weight")
plt.scatter(X_train[:,1],y_train)
plt.scatter(X_train[:,1],y_pred, color="red")
plt.xlabel("Vol")
plt.ylabel("Weight")
plt.grid(1)
plt.show()
```
